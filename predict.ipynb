## Making predictions 
> In this notebook we show how we can make predictions with our trained models.

### 1) Here we import the `'model'` module
> The `'model'` module has a predictor class to perform the following tasks:
> * Download historical prices and prepare the data for NN input.
> * Select the best trained model to make predicions.
> * Graph predictions.

```python
%load_ext autoreload
%aimport model
%autoreload 1

from model import Predictor
```

###  2) Select list of stocks to make predictions 
> Here we select a list of stock we want to make predictions on. We then download historical closing prices on these stocks and prepare the data for input into our trained models.

```python
#Our prediction list
# FB - Facebook
# KHC - The Kraft Heinz Company
# GOOG - Google 
# C - Citibank
tickers = ['SNAP', 'FB', 'EW', 'TSLA']
```

```python
#Select date range of closing prices
#YYYYMMDD date format
start_date = '20170701' 
end_date = '20190405'

#download quotes from yahoo and save to directory
for ticker in tickers:
    Predictor.download_prep(ticker,start_date,end_date)
```

### 3) Make predictions
 > * To make predictions we creat an instance of the Predictor class by passing a trade symbol.
 > * We then select the best model from all the trained models that best fits the historical trading prices of the current stock.
 > * After a best model is selected we can graph the prediction compared to the actual.

#### Make predictions for one stock
> Here we make predictions for one stock where print the test result for each trained model tested.

```python
stock_1 = tickers[0]
print("***  Closing price prediction for {} *** ".format(stock_1))
pred = Predictor(stock_1)
pred.select_model(verbose=1)
pred.graph()
```

#### Make predictions for the remaining stocks

```python
for ticker in tickers[1:]:
    print("***  Closing price prediction for {} *** ".format(ticker))
    pred = Predictor(ticker)
    pred.select_model(verbose=2)
    pred.graph()
```
